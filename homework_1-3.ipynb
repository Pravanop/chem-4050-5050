```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib
```

```python
#3(a)
h = 1
m = 1
a0 = 1 # bohr radii is 1 bohr
L = 1*a0
N = 2000
x = np.linspace(-L/2, L/2, N)
```

```python
#3(b)

I = np.eye(N)
I_off = np.eye(N, k = 1) + np.eye(N, k = -1)
# super and sub-diagonal is handled by k
laplacian = -2*I + I_off
delta_x = x[1] - x[0]
laplacian = laplacian/(delta_x**2)
```

```python
#3(c)

H = -(h**2/(2*m))*laplacian
```

```python
#3(d)

eigenvalues, eigenvectors = np.linalg.eigh(H)
sorted_eigenvalues_indices = np.argsort(eigenvalues) #sort the indices so that both eigenvaluees and eigenvectors are sorted
sorted_eigenvalues = eigenvalues[sorted_eigenvalues_indices]
sorted_eigenvectors = eigenvectors[:, sorted_eigenvalues_indices]
first_seven = sorted_eigenvalues[:7]
print("The first seven eigenvalues are:", first_seven)
```

```python
#five wavefunctions with their energy levels
matplotlib.rcParams.update({'font.size': 14})
fig, ax = plt.subplots(5, 1, figsize=(3.75, 3.75*5), sharex=True, sharey=True)
for i in range(5):
	ax[i].scatter(x, sorted_eigenvectors[i])
	ax[i].set_ylabel(f'$\psi_{i+1}$ (1/bohr)')
	ax[i].set_xlabel(f'x (bohr)')
plt.subplots_adjust(hspace=0.0, wspace=0.0)
```
