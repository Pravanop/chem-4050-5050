```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

Problem 1

```python
# read data
df = pd.read_csv('volume_pressure_data.csv')
df.head()
```

```python
#1(a)
#calculate mean, stddev, median

mean_vol = df['Volume'].mean()
mean_pressure = df['Pressure'].mean()
std_vol = df['Volume'].std()
std_pressure = df['Pressure'].std()
median_vol = df['Volume'].median()
median_pressure = df['Pressure'].median()

print(f"Mean Volume: {mean_vol}, Mean Pressure: {mean_pressure}")
print(f"Stddev Volume: {std_vol}, Stddev Pressure: {std_pressure}")
print(f"Median Volume: {median_vol}, Median Pressure: {median_pressure}")
```

```python
#1(b)
matplotlib.rcParams['figure.figsize'] = (3.75, 3.75)
matplotlib.rcParams['font.size'] = 12
plt.scatter(df['Volume'], df['Pressure'], color = 'gray', edgecolor = 'black', alpha = 0.7, s = 80, linewidth = 2)
plt.title('Volume and Pressure', fontsize = 12)
plt.grid(True)
plt.xlabel('Volume ($in^3$)')
plt.ylabel('Pressure (Hg)')
plt.show()
```

```python
#1(c)

fit = np.polyfit(df['Volume'], df['Pressure'], 2) # 2nd polynomial fit
x = np.linspace(min(df['Volume']), max(df['Volume']), 100)
y = np.polyval(fit, x) # calculates the y based on coefficients
plt.scatter(df['Volume'], df['Pressure'], color = 'gray', edgecolor = 'black', alpha = 0.7, s = 80, linewidth = 2, label = 'Raw Data')
plt.plot(x, y, color = 'red', linewidth = 2, label = 'Fit')
plt.title('Volume and Pressure with Fit', fontsize = 12)
plt.grid(True)
plt.xlabel('Volume ($in^3$)')
plt.ylabel('Pressure (Hg)')
plt.legend(frameon = False)
plt.show()
```

```python
#1(d)

#write down the function
print(f"Fitted function: P = {np.round(fit[0],2)}*(V**2) + {np.round(fit[1],2)}*V + {np.round(fit[2],2)}")
# P is inversiely proportional to V, so this is not making the most physical sense
```

```python
#1(e)

rmse = np.sqrt(np.mean((df['Pressure'] - np.polyval(fit, df['Volume']))**2))
print(f"RMSE: {np.round(rmse,3)}")
# the fit is not very good, especially at the edges of the data.
```
